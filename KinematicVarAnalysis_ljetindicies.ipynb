```python
# Load libraries
import numpy as np
from numpy import arange
from matplotlib import pyplot
from pandas import read_csv
from pandas import set_option
from pandas.plotting import scatter_matrix
```

```python
filename = 'jetindiciesttV.csv'
# choose names for columns
names = ['headers','LjetIndicies','Counted vRC jets','vRC jet vector size','top multiplicity']
dataset = read_csv(filename, delim_whitespace=False, names=names)
dataset=dataset.drop(['headers'], axis=1)
dataset
```

```python
dataset[(dataset['vRC jet vector size']==3) & (dataset['Counted vRC jets'] == dataset['vRC jet vector size'])]
```

```python
dataset[(dataset['vRC jet vector size']==3) & (dataset['Counted vRC jets'] != dataset['vRC jet vector size'])]
```

```python
#Fails mass cut on CRs
dataset[(dataset['Counted vRC jets'] > dataset['vRC jet vector size']) ]
```

```python
dataset[(dataset['Counted vRC jets'] < dataset['top multiplicity']) ]
```

```python
dataset[(dataset['Counted vRC jets'] < dataset['vRC jet vector size']) ]
```

```python
#Nothing surprising
dataset[(dataset['Counted vRC jets'] != dataset['LjetIndicies']) ]
```

```python
dataset[(dataset['Counted vRC jets'] == dataset['vRC jet vector size']) ]
```

```python
import matplotlib.pyplot as plt
# the histogram of the data
#n, bins, patches = plt.hist(dataset[(dataset['Counted vRC jets'])], 50, density=True, facecolor='g', alpha=0.75)

plt.hist(dataset.iloc[:,1:4],bins=[0,1,2,3,4,5,6,7,8]) 
plt.legend(['counted vRC jets','vRC jet vector size','tops'])
plt.yscale("log")
plt.title("vRC jet mulitplicity comparison") 
plt.show()
```
