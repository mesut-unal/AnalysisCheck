```python
# Load libraries
import numpy
from numpy import arange
from matplotlib import pyplot
from pandas import read_csv
from pandas import set_option
from pandas.plotting import scatter_matrix
```

```python
filename = 'metmZtHTcheck.csv'
# choose names for columns
names = ['headers','MET','mZt','HT','mZt-HT','topflag','mass','pt','Njet','Nconst','Ntops','vrc_maxM_pre0']
dataset = read_csv(filename, delim_whitespace=False, names=names)
dataset=dataset.drop(['headers'], axis=1)
dataset.sort_values('vrc_maxM_pre0')
```

```python
#failures
dataset[(dataset['MET']>=dataset['mZt-HT']) ].sort_values('vrc_maxM_pre0').sort_values('Ntops')
```

```python
#failures
dataset[(dataset['MET']>=dataset['mZt-HT']) & (dataset['vrc_maxM_pre0']<140.)]
```

```python
#survivors
dataset[(dataset['MET']<dataset['mZt-HT'])].sort_values('vrc_maxM_pre0').sort_values('Ntops')
```

```python
#fails on meff cut, survives Njets>=1
dataset[(dataset['MET']>=dataset['mZt-HT']) & (dataset['Njet']>0)].sort_values('vrc_maxM_pre0')
```

```python
#RESULT for ttbarV: There is no vRC jet leading mass <140 GeV events at all. 
```

```python
dataset[(dataset['Njet']>0) & (dataset['MET']<dataset['mZt-HT'])].sort_values('Ntops')
```

```python
#Confirms previous cell: Survivors are all top events
```
